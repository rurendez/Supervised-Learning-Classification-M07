## Exercise 1: Machine Learning model creation
    
The data provided for this exercise refer to wine recognition data. For this data we have to create at least two different classification models in order to best predict the classes on given file.

### File column description

```python
# Let's display the contents of the data file info about our wine recognintion datset

infofile = open('wineNames.txt', 'r')
file_contents = infofile.read()
print (file_contents)
infofile.close()
```

### Load and Explore Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

# Import data from file. Notice that fields are separated by "::", with no header
colnames = ['Class', 'Alcohol', 'Malic_Acid', 'Ash', 'Ash_Alcanity', 'Magnesium',
       'Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols',
       'Proanthocyanins', 'Color_Intensity', 'Hue', 'OD280', 'Proline']

# Read the csv file
wine_df = pd.read_csv("wineData.txt", sep=",", encoding='latin-1', header=None, engine='python', 
                   names=colnames, index_col=False)
wine_df.sample(10).T
```

```python
# Now, let's show the data types and non-null values. As we can see the dataset is already cleaned, 
# there are no missing values
wine_df.info()
```

```python
 # Check for zeros in all columns 
wine_df[wine_df == 0].count()   
```

```python
# Take a picture of the statistical data of dataframe
wine_df.describe().T
```

```python
# Visualize each variable's boxplot in order to check the outliers
from scipy import stats

fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(12, 8))
index = 0
axs = axs.flatten()
for k,v in wine_df.items():
    sns.boxplot(y=k, data=wine_df, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

As we can see on these boxplot charts there are not too many outliers. Only on a few attributes.

```python
# Visualize each variable's distribution
wine_df.hist(bins=50, figsize=(16,12))
plt.rcParams['figure.figsize'] = [8, 4]
plt.show()
```

As we can see in all these charts, the attributes do not follow a Normal Distribution. They do not have the bell shape.

```python
# Now let's display de correlation graph
corr = wine_df.corr()
plt.figure(figsize=(16,16))
plt.title('Correlation graph', fontsize=30)
sns.set(font_scale=1.4)
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
# Let's display how many classes we have from each group
sns.countplot(x = wine_df['Class'])
```

### Data Preparation

```python
# Let's split features, and target
X = wine_df.drop(['Class'],axis=1)
y = wine_df['Class']
```

```python
# Let's split our dataset into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

### Classification Models

```python
#importing different classification models
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score,f1_score
from sklearn.metrics import classification_report
```

```python
# Let's create a table for model comparison
model_compare = {}
```

```python
# Decision Tree
dt_classifier = DecisionTreeClassifier(random_state = 0)
dt_classifier.fit(X_train,y_train)
y_pred_dt = dt_classifier.predict(X_test)

print(f"Model Accuracy : {accuracy_score(y_pred_dt,y_test)*100:.2f}%")
print(f"Model F1-Score : {f1_score(y_pred_dt,y_test,average='weighted')*100:.2f}%")

dt_accuracies = cross_val_score(estimator = dt_classifier, X = X_train, y = y_train, cv = 5)

print("Cross Val Accuracy: {:.2f} %".format(dt_accuracies.mean()*100))
print("Cross Val Standard Deviation: {:.2f} %".format(dt_accuracies.std()*100))
print(classification_report(y_pred_dt,y_test,zero_division=1))

model_compare['Decision Tree']=[accuracy_score(y_pred_dt,y_test),f1_score(y_pred_dt,y_test,average='weighted'),
                                (dt_accuracies.mean()),(dt_accuracies.std())]
```

```python
# KNN Classifier
kn_classifier = KNeighborsClassifier()
kn_classifier.fit(X_train,y_train)
y_pred_kn = kn_classifier.predict(X_test)

print(f"Model Accuracy : {accuracy_score(y_pred_kn,y_test)*100:.2f}%")
print(f"Model F1-Score : {f1_score(y_pred_kn,y_test,average='weighted')*100:.2f}%")

kn_accuracies = cross_val_score(estimator = kn_classifier, X = X_train, y = y_train, cv = 5)

print("Cross Val Accuracy: {:.2f} %".format(kn_accuracies.mean()*100))
print("Cross Val Standard Deviation: {:.2f} %".format(kn_accuracies.std()*100))
print(classification_report(y_pred_kn,y_test,zero_division=1))

model_compare['KNN']=[accuracy_score(y_pred_kn,y_test),f1_score(y_pred_kn,y_test,average='weighted'),
                      (kn_accuracies.mean()),(kn_accuracies.std())]
```

```python
# SVM  
sv_classifier = SVC()
sv_classifier.fit(X_train,y_train)
y_pred_sv = sv_classifier.predict(X_test)

print(f"Model Accuracy : {accuracy_score(y_pred_sv,y_test)*100:.2f}%")
print(f"Model F1-Score : {f1_score(y_pred_sv,y_test,average='weighted')*100:.2f}%")

sv_accuracies = cross_val_score(estimator = sv_classifier, X = X_train, y = y_train, cv = 5)

print("Cross Val Accuracy: {:.2f} %".format(sv_accuracies.mean()*100))
print("Cross Val Standard Deviation: {:.2f} %".format(sv_accuracies.std()*100))
print(classification_report(y_pred_sv,y_test,zero_division=1))

model_compare['Support Vector Classifier']=[accuracy_score(y_pred_sv,y_test),f1_score(y_pred_sv,y_test,average='weighted'),
                                            (sv_accuracies.mean()),(sv_accuracies.std())]
```

Train the models using allowed parameteres in order to improve prediction.

```python
# Decision Tree
dt_classifier=DecisionTreeClassifier(splitter = "random", random_state = 0)
dt_classifier.fit(X_train,y_train)
y_pred_dt = dt_classifier.predict(X_test)

print(f"Model Accuracy : {accuracy_score(y_pred_dt,y_test)*100:.2f}%")
print(f"Model F1-Score : {f1_score(y_pred_dt,y_test,average='weighted')*100:.2f}%")

dt_accuracies = cross_val_score(estimator = dt_classifier, X = X_train, y = y_train, cv = 5)

print("Cross Val Accuracy: {:.2f} %".format(dt_accuracies.mean()*100))
print("Cross Val Standard Deviation: {:.2f} %".format(dt_accuracies.std()*100))
print(classification_report(y_pred_dt,y_test,zero_division=1))

model_training_compare['Decision Tree']=[accuracy_score(y_pred_dt,y_test),f1_score(y_pred_dt,y_test,average='weighted'),
                                (dt_accuracies.mean()),(dt_accuracies.std())]
```

```python
# KNN Classifier
kn_classifier = KNeighborsClassifier(n_neighbors = 1)
kn_classifier.fit(X_train,y_train)
y_pred_kn = kn_classifier.predict(X_test)

print(f"Model Accuracy : {accuracy_score(y_pred_kn,y_test)*100:.2f}%")
print(f"Model F1-Score : {f1_score(y_pred_kn,y_test,average='weighted')*100:.2f}%")

kn_accuracies = cross_val_score(estimator = kn_classifier, X = X_train, y = y_train, cv = 5)

print("Cross Val Accuracy: {:.2f} %".format(kn_accuracies.mean()*100))
print("Cross Val Standard Deviation: {:.2f} %".format(kn_accuracies.std()*100))
print(classification_report(y_pred_kn,y_test,zero_division=1))

model_training_compare['KNN']=[accuracy_score(y_pred_kn,y_test),f1_score(y_pred_kn,y_test,average='weighted'),
                      (kn_accuracies.mean()),(kn_accuracies.std())]
```

In this model KNN forcing n_neighbors to 1, we have manage to increase the accuracy from 73.33% to 75.56%

```python
# SVM  
sv_classifier = SVC(kernel = 'linear', random_state = 0)
sv_classifier.fit(X_train,y_train)
y_pred_sv = sv_classifier.predict(X_test)

print(f"Model Accuracy : {accuracy_score(y_pred_sv,y_test)*100:.2f}%")
print(f"Model F1-Score : {f1_score(y_pred_sv,y_test,average='weighted')*100:.2f}%")

sv_accuracies = cross_val_score(estimator = sv_classifier, X = X_train, y = y_train, cv = 5)

print("Cross Val Accuracy: {:.2f} %".format(sv_accuracies.mean()*100))
print("Cross Val Standard Deviation: {:.2f} %".format(sv_accuracies.std()*100))
print(classification_report(y_pred_sv,y_test,zero_division=1))

model_training_compare['Support Vector Classifier']=[accuracy_score(y_pred_sv,y_test),f1_score(y_pred_sv,y_test,average='weighted'),
                                            (sv_accuracies.mean()),(sv_accuracies.std())]
```

In this model SVM we have added the parameter kernel='linear', and we have manage to increase the accuracy from 82.22% to 97.78%

Apply some engineering process to improve the results (normalization, standardization, sampling...).

### Data Balancing - SMOTE

```python
# Let's balance de dataset using SMOTE in order to obtain a better model 
from imblearn.over_sampling import SMOTE
sm=SMOTE()
X,y=sm.fit_resample(X,y)
```
